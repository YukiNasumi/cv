```python
from torchvision import transforms
from matplotlib import pyplot as plt
import numpy as np
import PIL
from PIL import Image
import torch
image = Image.open('QQ图片20240712184923.jpg')
type(image)
```

```python
def show_tensor(image, plot=False):
    '''把image对象或tensor对象变成图片'''
    if isinstance(image, torch.Tensor) or isinstance(image,np.ndarray): #or isinstance(image,PIL.JpegImagePlugin.JpegImageFile):
        if not image.shape[-1]==3:
            trans = transforms.ToPILImage()#这里是把图像转换为channel后置的类型
            im = trans(image)
            plt.imshow(im)
            plt.axis('off')
            plt.show()
        plt.imshow(image)
        plt.axis('off')
        plt.show()
    else:
        if plot == True:
            trans1 = transforms.ToTensor()
            trans2 = transforms.ToPILImage()
            im = trans2(trans1(image))
            plt.imshow(im)
            plt.axis('off')
            plt.show()
            return
        image.show()
        
```

```python
width, height = image.size
print(width,height)
```

```python

show_tensor(image,plot=True)
trans1 = transforms.Resize((2,5))
show_tensor(trans1(image),plot=True)
center_crop = transforms.CenterCrop((100, 256))  
show_tensor(center_crop(image),plot=True)
```
