```python
from torchvision import transforms
from PIL import Image
import torch

# 打开图像文件
image = Image.open('QQ图片20240712184923.jpg')

# 定义转换操作  
transform = transforms.ToTensor()

# 将图像转换为张量
tensor_image = transform(image)
tensor_image.shape
```

```python
transform2 = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.CenterCrop(128),
    transforms.ToTensor()
])
```

```python
tensor_image2 = transform2(image)
#tensor_image2.shape
tensor_image2.shape
```

```python
from torchvision import transforms
import matplotlib.pyplot as plt

# 创建一个示例张量
tensor_image = tensor_image2

# 定义转换操作
transform = transforms.ToPILImage()

# 将张量转换为图像
image = transform(tensor_image2)


# 或使用 matplotlib 显示图像
plt.imshow(image)
plt.axis('off')
plt.show()
```
